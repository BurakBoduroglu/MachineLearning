**Missing Values** - **Simple Imputer**

```python
age = datas.iloc[:,1:4].values # iloc[row begin : row finish , column begin : column finish]
```

```python
age
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = "mean")
```

```python
imputer = imputer.fit(age)       # 2D or more dimensional array need.
age = imputer.transform(age)     # fit and then transform it.
print(age)
```
